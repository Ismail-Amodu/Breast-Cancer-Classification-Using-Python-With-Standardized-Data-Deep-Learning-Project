```python
# Importing dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

**Data collection and processing**

```python
# Loading data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# Loading dataset into pandas dataframe
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Printing the first-five rows of the dataframe
data_frame.head()
```

```python
# Adding the 'target' column to the dataframe
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Printing the last-five rows of the dataframe
data_frame.tail()
```

```python
# Checking the number of rows and columns in the dataframe
data_frame.shape
```

```python
# Getting some information about the dataframe
data_frame.info()
```

```python
# Checking whether there are missing values
data_frame.isnull().sum()
```

```python
# Obtaining the statistical measures of the dataframe
data_frame.describe()
```

```python
# Obtaining the distribution of target variable
data_frame['label'].value_counts()
```

1 --> Benign
0 --> Malignant

```python
# Obtaining the mean of the target variables
data_frame.groupby('label').mean()
```

```python
# Separating the features and target
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Splitting dataset into training and test dataset
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=5)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
!pip install scikit-learn --upgrade
```

```python
# Data standardization
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train)
X_test_std = scaler.transform(X_test)
print(X_train_std)
```

```python
# Building Neural Network
import tensorflow as tf
tf.random.set_seed(5)
from tensorflow import keras
```

```python
# Setting up the layers of Neural Network
model = keras.Sequential([
                            keras.layers.Flatten(input_shape=(30,)),
                            keras.layers.Dense(20, activation='relu'),
                            keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the Neural Network
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training the Neural Network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=12)
```

```python
# VISUALIZATION
# Visualizing the accuracy and loss
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc='upper right')
```

```python
# Accuracy on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
print(X_test_std.shape)
```

```python
Y_pred = model.predict(X_test_std)
print(Y_pred.shape)
print(Y_pred[0])
```

```python
print(X_test_std)
print(Y_pred)

# Note that model.predict() gives the prediction probability of each class for that data point
```

```python
# Conversion from prediction probability to class label
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```

Building a predictive system

```python
# Building a predictive system
input_data = (819.17,24.8,132.4,1123,0.0974,0.2458,0.2065,0.1118,0.2397,0.078,0.9555,3.568,11.07,116.2,0.003139,0.08297,0.0889,0.0409,0.04484,0.01284,20.96,29.94,151.7,1332,0.1037,0.3903,0.3639,0.1767,0.3176,0.1023)

# Changing input data into numpy array
input_data_as_numpy_array = np.asarray(input_data)

# Reshape the numpy array for one instance prediction
input_data_reshaped = input_data_as_numpy_array.reshape(1, -1)

# Standardinsing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if (prediction_label[0] == 0):
    print('The tumor is Malignant')
else:
    print('The tumor is Benign')
```
